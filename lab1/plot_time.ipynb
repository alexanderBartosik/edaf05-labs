```python
import matplotlib.pyplot as plt
import numpy as np

def extract_data_from_file(filename):
    N = []  # Vector for N values
    t = []  # Vector for time values

    with open(filename, 'r') as file:
        for line in file:
            # Splitting each line by commas to separate the fields
            fields = line.split(',')
            # Extracting N value
            n_value = int(fields[0].split(':')[1].strip())
            # Extracting time value
            time_value = float(fields[1].split(':')[1].strip())

            # Append the extracted values to the lists
            N.append(n_value)
            t.append(time_value)

    return N, t

# Usage
filename = "./runtime.txt"
N, t = extract_data_from_file(filename)
print("N values:", N)
print("time values:", t)

# Creating the plot
plt.figure(figsize=(10, 6))  # Set the figure size for better readability
plt.plot(N, t, 'o-', label='Runtime vs N')  # Plot data with points connected by lines
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.yscale('log')  # Set x-axis to logarithmic scale
plt.xlabel('N (log scale)')  # X-axis label
plt.ylabel('Time (seconds)')  # Y-axis label
plt.title('Runtime Analysis on Log Scale')  # Title of the plot
plt.legend()  # Add a legend
plt.grid(True)  # Enable grid for better readability of log scale
plt.show()  # Display the plot



plt.figure(figsize=(10, 6))  # Set the figure size for better readability

# Perform linear regression
coefficients = np.polyfit(np.log(N), np.log(t), 1)
m = coefficients[0]  # Slope
b = coefficients[1]  # Intercept

# Generate predicted values
y_pred = m * np.log(N) + b

# Plot the data and regression line
plt.scatter(np.log(N), np.log(t), color='blue', label='Actual')
plt.plot(np.log(N), y_pred, color='red', label='Regression Line')
plt.xlabel('log N')
plt.ylabel('log time')
plt.title(f'Linear Regression: Slope (k) = {m:.2f}')
plt.legend()
plt.show()

print(m)
```

```python
y = lambda x: x**2
xx = np.linspace(1, 1000, 100)

yy = y(xx) 

h = 1e-5
deriv = lambda f, x: (f(x + h) - f(x)) / h
# To compute the derivative of y = x^2, pass the function y itself to the derivative function
yprime = deriv(y, xx)  # Now y is a function, which is correct

plt.figure()
plt.plot(xx, yy, label='plot of y=x^2')
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.legend()
plt.show()

plt.figure()
plt.plot(xx, yprime, label="Derivative of y=x^2")
plt.xlabel('x')
plt.ylabel("y'")
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Perform linear regression
coefficients = np.polyfit(x, y, 1)
m = coefficients[0]  # Slope
b = coefficients[1]  # Intercept

# Generate predicted values
y_pred = m * x + b

# Plot the data and regression line
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
